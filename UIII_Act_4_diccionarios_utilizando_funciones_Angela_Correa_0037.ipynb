```python
# 1. Diccionario principal
relojeros = {}

# 2. Definición de Funciones
def agregar_relojero():
    nombre = input("Nombre Relojero: ")
    herramientas = input("Herramientas: ")
    marcas = input("Marcas Lujo: ")
    stock = input("Nivel Stock: ")
    complejas = input("Rep. Complejas: ")
    garantia = input("Garantía: ")
    taller = input("Taller: ")

    relojeros[nombre] = {
        "Herr": herramientas,
        "Marc": marcas,
        "Stck": stock,
        "Cmpl": complejas,
        "Garn": garantia,
        "Tall": taller
    }
    print("-> Guardado con éxito.")

def listar_relojeros():
    if not relojeros:
        print("\n[!] El taller está vacío.")
        return

    # Tabla compacta (Distribución optimizada)
    print("\n" + "="*95)
    print(f"{'Nombre':<12} | {'Herr':<12} | {'Marc':<12} | {'Stck':<6} | {'Cmpl':<8} | {'Garn':<8} | {'Tall':<12}")
    print("-" * 95)

    # Uso de for clave, valor (key, value)
    for nombre, r in relojeros.items():
        print(f"{nombre[:12]:<12} | "
              f"{r['Herr'][:12]:<12} | "
              f"{r['Marc'][:12]:<12} | "
              f"{r['Stck'][:6]:<6} | "
              f"{r['Cmpl'][:8]:<8} | "
              f"{r['Garn'][:8]:<8} | "
              f"{r['Tall'][:12]:<12}")
    print("=" * 95)

def buscar_relojero():
    nombre = input("Relojero a buscar: ")
    if nombre in relojeros:
        print(f"\nFicha técnica de {nombre}:")
        for k, v in relojeros[nombre].items():
            print(f" - {k}: {v}")
    else:
        print("No se encontró el registro.")

def borrar_relojero():
    nombre = input("Nombre a borrar: ")
    if nombre in relojeros:
        del relojeros[nombre]
        print("Eliminado.")
    else:
        print("No existe.")

def actualizar_relojero():
    nombre = input("Nombre a actualizar: ")
    if nombre in relojeros:
        relojeros[nombre]['Stck'] = input("Nuevo Stock: ")
        relojeros[nombre]['Garn'] = input("Nueva Garantía: ")
        print("Datos actualizados.")
    else:
        print("Relojero no encontrado.")

# --- INFOGRAFÍA DEL PROYECTO ---
# Estudiante: Angela Michelle Correa Lopez
# Especialidad: Programación
# Plantel: CBTIS No. 128
# Docente: Eliseo Nava
# Proyecto: Sistema de Gestión de Relojería de Lujo 2026

# --- MOTOR PRINCIPAL ---
while True:
    print("\n>>> SISTEMA RELOJERO <<<")
    print("1.Agregar 2.Listar 3.Buscar 4.Borrar 5.Actualizar 6.Salir")
    opcion = input("Seleccione la opcion: ")

    if opcion == "1":
        agregar_relojero()
    elif opcion == "2":
        listar_relojeros()
    elif opcion == "3":
        buscar_relojero()
    elif opcion == "4":
        borrar_relojero()
    elif opcion == "5":
        actualizar_relojero()
    elif opcion == "6":
        print("Cerrando taller...")
        break
    else:
        print("Opción inválida.")
print("Programa realizado por Angela Michelle Correa Lopez NL = 13")
```
